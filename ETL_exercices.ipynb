### Ex E5 ⭐⭐ — Fonction extract()
Encapsulez tout dans une fonction `extract(q1_path, q2_path, clients_path, produits_path) -> dict`
qui retourne un dictionnaire avec les 4 DataFrames harmonisés et consolidés.
La fonction doit afficher des logs à chaque étape.

```python
import pandas as pd
import logging 

# Configuration de base :
logging.basicConfig(level=logging.INFO,format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",datefmt="%H:%M:%S",force=True,)
logger = logging.getLogger("ETL")

q1_path= "commandes_Q1.csv"
q2_path= "commandes_Q2.csv"
clients_path= "clients.csv"
produits_path= "produits.csv"


def extract(q1_path, q2_path, clients_path, produits_path) -> dict:

    logging.info("Chargement des fichiers CSV")
    df_q1 = pd.read_csv(q1_path)
    df_q2 = pd.read_csv(q2_path)
    df_clients = pd.read_csv(clients_path)
    df_produits = pd.read_csv(produits_path)

    logging.info("Rename de la colonne 'moyen_paiement'")
    df_q2 = df_q2.rename(columns={"moyen_paiement":"mode_paiement"})

    logging.info("Ajout de la colonne devise dans Q1")
    df_q1["devise"] = "EUR"

    logging.info("Ajout de la colonne 'source'")
    df_q1["source"] = "Q1_2024"
    df_q2["source"] = "Q2_2024"

    logging.info("Concat Q1 et Q2")
    df_commandes = pd.concat([df_q1, df_q2], ignore_index=True)
  
    logging.info("Création d'un dictionnaire")
    data = {
        "commandes": df_commandes,
        "clients": df_clients,
        "produits": df_produits,
    }

    return data

resultat = extract(q1_path, q2_path, clients_path, produits_path)
```

```python
# 2. `saison` : depuis le mois (Hiver/Printemps/Été/Automne)

def saison(mois):
    if mois in [12, 1, 2]:
        return 'Hiver'
    elif mois in [3, 4, 5]:
        return 'Printemps'
    elif mois in [6, 7, 8]:
        return 'Été'
    else:
        return 'Automne'

df_commandes['saison'] = df_commandes['mois'].apply(saison)
```

### Ex T8 ⭐⭐⭐ — Validation qualité
Écris une fonction `valider_pipeline(df) -> bool` qui :
1. Vérifie qu'il n'y a plus de NaN sur les colonnes critiques
2. Vérifie qu'il n'y a plus de doublons
3. Vérifie que tous les prix sont > 0
4. Vérifie que toutes les quantités sont > 0
5. Vérifie que montant_total == quantite × prix_unitaire (à 1 centime près)
6. Retourne True si tout est OK, False sinon avec les erreurs détaillées

```python
def valider_pipeline(df) -> bool :

    erreurs = []

    # Verif NaN
    if df[['montant_total', 'statut', 'date_commande']].isna().sum().sum() > 0 :
        erreurs.append("NaN détectés")

    # Verif Doublons
    if df.duplicated().sum() > 0 :
        erreurs.append("Doublons détectés")

    # Verif prix positif
    if (df['prix_unitaire'] <= 0).sum() > 0:
        erreurs.append("Prix négatifs détectés")


    # Verif quantité positive
    if (df['quantite'] <= 0).sum() > 0:
        erreurs.append("Quantités négatives détectées")


    # Verif montant_total == quantite x prix_unitaire
    if (~np.isclose(df['montant_total'], df['quantite'] * df['prix_unitaire'], atol=0.01)).sum() > 0:
        erreurs.append("Incohérences sur montant_total")

    # Résultats de la validation qualités
    if erreurs:                   
        for e in erreurs:          
            print(e)               
        return False  

    print("Pipeline valide")
    
    return True
    
```

```python
# Fonction Transform

def transform(sources : dict) -> pd.DataFrame:

    """Nettoie, enrichit et fusionne les tables commandes, clients et produits.

    Parameters
    ----------
    sources : dict
        Dictionnaire contenant les DataFrames 'commandes', 'clients' et 'produits'.

    Returns
    -------
    pd.DataFrame
        Table consolidée et nettoyée.
    """

    logger.info("Début de transformation")

    # On extrait les csv du dictionnaire
    df_cmd = sources["commandes"].copy()
    df_cli = sources["clients"].copy()
    df_pro = sources["produits"].copy()

    # On supprime les doublons de commandes
    df_cmd = df_cmd.drop_duplicates()
    logger.info("Doublons supprimés dans commandes")

    # Date
    df_cmd["date_commande"] = pd.to_datetime(df_cmd["date_commande"], format='ISO8601')
    df_cmd["annee"] = df_cmd["date_commande"].dt.year
    df_cmd["mois"] = df_cmd["date_commande"].dt.month
    df_cmd["trimestre"] = df_cmd["date_commande"].dt.quarter
    df_cmd["jour_semaine"] = df_cmd["date_commande"].dt.day_name()
    logger.info("Conversion en datetime et nouvelles colonnes (annee, mois, trimestre et jours)")

    # 'quantite en numerique
    df_cmd['quantite'] = df_cmd['quantite'].astype(int)

    # On supprime les prix négatifs et quantités nulles ou négatives
    df_cmd = df_cmd[df_cmd["prix_unitaire"] > 0]
    df_cmd = df_cmd[df_cmd["quantite"] > 0]
    logger.info("Suppression des prix et quantités négatifs")   

    # Gestion des NaN
    df_cmd["statut"] = df_cmd["statut"].fillna("inconnu")
    df_cmd["montant_total"] = df_cmd["montant_total"].fillna(df_cmd["quantite"] * df_cmd["prix_unitaire"])
    logger.info("Remplacement des NaNpar des valeurs")

    # On supprime les doublons dans clients
    df_cli = df_cli.drop_duplicates()
    logger.info("Suppression des doublons dans clients")

    # On garde que les produits actifs
    df_pro = df_pro[df_pro['actif'] == True]
    logger.info("Filtrage des produits actifs")

    # Jointures 
    df_cmd = pd.merge(df_cmd,
         df_cli[['client_id', 'ville', 'region', 'segment', 'date_inscription']],
         how="left",
         on = "client_id")
    logger.info("Jointure del table clients")

    df_cmd = pd.merge(df_cmd,
         df_pro[['produit_id', 'nom', 'categorie', 'fournisseur']],
         how="left",
         on = "produit_id")
    logger.info("Jointure del table produits")

    # On rajoute la colonne tranche_montant
    df_cmd['tranche_montant'] = pd.cut(
    df_cmd['montant_total'],
    bins=[0, 100, 500, 2000, 10000],       
    labels=['Petit', 'Moyen', 'Grand', 'Très grand'])
    logger.info("Ajout de la colonne tranche_montant")


    # On rajoute la colonne saison
    def saison(mois):
        if mois in [12, 1, 2]:
            return 'Hiver'
        elif mois in [3, 4, 5]:
            return 'Printemps'
        elif mois in [6, 7, 8]:
            return 'Été'
        else:
            return 'Automne'

    df_cmd['saison'] = df_cmd['mois'].apply(saison)
    logger.info("Ajout de la colonne saison")

    # On rajoute la colonne est_annulée
    df_cmd['est_annulee'] = df_cmd['statut'] == 'annulée'
    logger.info("Ajout de la colonne 'est_annulée'")

    # On rajoute la colonne ca_perdu
    df_cmd['ca_perdu'] = np.where(df_cmd['est_annulee'], df_cmd['montant_total'], 0)
    logger.info("Ajout de la colonne 'ca_perdu'")

    # On rajoute la colonne date_inscription_commande
    df_cmd["date_inscription"] = pd.to_datetime(df_cmd["date_inscription"], format='ISO8601')
    df_cmd['delai_inscription_commande'] = (df_cmd['date_commande'] - df_cmd['date_inscription']).dt.days
    logger.info("Ajout de la colonne 'date_inscription_commande'")
    
    return df_cmd
```
